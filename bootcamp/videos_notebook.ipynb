```python
url = "C:/University/CS/Image_Processing/bootcamp/video1_category1.mp4"
vid = media.read_video(url)
print("shape:",vid.shape,"data type:", vid.dtype)
print(vid.metadata)
```

```python
number_of_frames = vid.shape[0]
frame = vid[10]
```

```python
plt.imshow(frame)
plt.show()
```

```python
frame.shape
```

```python
gray_frame = frame @ [0.299, 0.587, 0.114]
```

```python
plt.imshow(gray_frame, cmap='gray')
plt.show()
```

```python
gray_vid = []
```

```python
for i in range(number_of_frames):
    gray_vid.append(vid[i] @ [0.299, 0.587, 0.114])
gray_vid = np.array(gray_vid)
```

mean, sum, std, and max value

```python
rgb_pre_frame_mean = []
rgb_pre_frame_sum = []
rgb_pre_frame_std = []
rgb_pre_frame_max = []
gray_pre_frame_mean = []
gray_pre_frame_sum = []
gray_pre_frame_std = []
gray_pre_frame_max = []
```

```python
for frame in range(number_of_frames):
    rgb_pre_frame_mean.append(np.mean(vid[frame]))
    rgb_pre_frame_sum.append(np.sum(vid[frame]))
    rgb_pre_frame_std.append(np.std(vid[frame]))
    rgb_pre_frame_max.append(np.max(vid[frame]))
    gray_pre_frame_mean.append(np.mean(gray_vid[frame]))
    gray_pre_frame_sum.append(np.sum(gray_vid[frame]))
    gray_pre_frame_std.append(np.std(gray_vid[frame]))
    gray_pre_frame_max.append(np.max(gray_vid[frame]))
```

```python
import matplotlib.pyplot as plt

# Assuming you have already calculated the lists rgb_pre_frame_mean, rgb_pre_frame_sum, ...
# and gray_pre_frame_mean, gray_pre_frame_sum, ...

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot mean values
axs[0, 0].plot(rgb_pre_frame_mean, label='RGB')
axs[0, 0].plot(gray_pre_frame_mean, label='Grayscale')
axs[0, 0].set_title('Mean Values')
axs[0, 0].legend()

# Plot sum values
axs[0, 1].plot(rgb_pre_frame_sum, label='RGB')
axs[0, 1].plot(gray_pre_frame_sum, label='Grayscale')
axs[0, 1].set_title('Sum Values')
axs[0, 1].legend()

# Plot standard deviation values
axs[1, 0].plot(rgb_pre_frame_std, label='RGB')
axs[1, 0].plot(gray_pre_frame_std, label='Grayscale')
axs[1, 0].set_title('Standard Deviation Values')
axs[1, 0].legend()

# Plot maximum values
axs[1, 1].plot(rgb_pre_frame_max, label='RGB')
axs[1, 1].plot(gray_pre_frame_max, label='Grayscale')
axs[1, 1].set_title('Maximum Values')
axs[1, 1].legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
